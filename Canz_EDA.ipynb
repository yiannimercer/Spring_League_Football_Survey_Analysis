### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.options.display.max_rows = 999
```

```python
df = pd.read_excel('TSL Survey draft 1_April 26, 2021_18.06.xlsx')
```

```python
#correcting the headers
```

```python
df.columns = df.iloc[0]
df = df[1:]
```

```python
df
```

## Cleaning Data

```python
#df shape
df.shape
```

```python
df.isnull().sum()
```

```python
# Considering there are 164 observations, we can remove the columns " When you watch pro football what 
                                                                    # do you enjoy most about it? - other - Text"
# because only 1 person answered the question

df.drop('When you watch pro football what do you enjoy most about it? - other - Text',inplace = True, axis = 1)
```

```python
#lets examine the 'Are you interested in pro football outside of the NFL?' column as we have a lot of missing
#values, but not as many as as the most 
df['Are you interested in pro football outside of the NFL?'].value_counts()

#seems to be a multiple choice question that respondents were allowed to skip
#opt to remove the column as significant number of people did not answer it and the remaining values do not
#represent the population (at least we are unable to make the claim that they do)
```

```python
#removing the 'Are you interested in pro football outside of the NFL?' column
df.drop('Are you interested in pro football outside of the NFL?',inplace = True, axis = 1)
```

```python
#% of missing valuaes per column
df.isnull().sum()/df.shape[0]
```

```python
#before we go into any more data cleaning by column, let's examine the missing values by rows as these rows 
#represent each individual person completing the survey

missing_rows = df.isnull().sum(axis = 1)
missing_rows.sort_values(ascending = False).head(40)

##with 15 columns/questions, row 11 is missing everything --> we will remove 
```

```python
df.drop(11, inplace = True)
```

```python
##Cleaning up "numeric" column(s)
df.isnull().sum()

#Income question is missing 2 values.  Let's find out the mode (its multiple choice) and fill in that option for
#2 missing values
```

```python
#Filling in income column with the mode
df['Income, estimation in thousands of USD'].fillna(df['Income, estimation in thousands of USD'].mode()[0],inplace= True)
df.isnull().sum()
```

```python
#same as above for the below column --> the overwhelming majority watch football for entertainment.  
df['When you watch pro football what do you enjoy most about it? - Selected Choice'].fillna(df['When you watch pro football what do you enjoy most about it? - Selected Choice'].mode()[0],inplace = True)
df.isnull().sum()
```

```python
#I will now explore the percentages of completion per individual 
#and remove rows where they differ signficantly from the mean
missing_rows = df.isnull().sum(axis = 1)
missing_rows.sort_values(ascending = False)

#I will remove respondants who did not fill out 40% or more of the questions
```

```python
#index of rows that are missing 40% or more 
x = np.where(df.isnull().mean(axis=1) > 0.4)
x = np.asarray(x)
x = x.tolist()
missing_index = []
for i in x:
    for a in i:
        missing_index.append(a)
missing_index
```

```python
df_cleaned = df.drop(missing_index,axis = 0)
```

```python
df_cleaned.isnull().sum()
```

### Data has been thoroughly cleaned without severely affecting the actual data that is being communicated (e.g., filling NA values that may not reflect the sample, and in turn population)
